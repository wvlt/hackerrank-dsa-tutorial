```python
class SinglyLinkedListNode:
    def __init__(self, node_data):
        self.data = node_data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
```

## ğŸ’¡ Solution: Insert at Head

**Time Complexity:** O(1) - Constant time operation!  
**Space Complexity:** O(1) - Only create one new node

**Key Steps:**
1. Create a new node with the given data
2. Link the new node's next to the current head
3. Return the new node as the new head

**Why is this O(1)?** We don't need to traverse - we just update pointers!


```python
def insertNodeAtHead(llist, data):
    """
    Insert a node at the head of a linked list.
    
    Args:
        llist: Pointer to the head node (can be None)
        data: Integer value for the new node
        
    Returns:
        Pointer to the new head node
    """
    new_node = SinglyLinkedListNode(data)
    
    # Link new node to the current head (works even if head is None)
    new_node.next = llist
    
    # Return the new node as the new head
    return new_node

# Test the function
llist = SinglyLinkedList()
llist.head = insertNodeAtHead(None, 13)
llist.head = insertNodeAtHead(llist.head, 16)
llist.head = insertNodeAtHead(llist.head, 15)

print("Linked list after insertions:")
current = llist.head
while current:
    print(current.data, end=" -> ")
    current = current.next
print("NULL")
```

## ğŸ” **Step-by-Step Walkthrough**

**Example:** Insert 15 into `16 -> 13 -> NULL`

1. **Create new node:** `new_node` with data 15
2. **Link to current head:** `new_node.next = llist` (points to 16)
3. **Return new head:** Return `new_node` (15)
4. **Result:** `15 -> 16 -> 13 -> NULL`

**Note:** This works even if the list is empty (llist is None)!

---

## ğŸ” **Complexity Analysis**

- **Time Complexity:** O(1) - Constant time! No traversal needed
- **Space Complexity:** O(1) - We only create one new node

**Comparison with Tail Insertion:**
- Head insertion: O(1) - Much faster!
- Tail insertion: O(n) - Must traverse to end

---

## ğŸ¯ **Key Takeaways**

1. **Head insertion is O(1):** No traversal needed, just pointer updates
2. **Works with empty list:** `new_node.next = None` when list is empty
3. **Return new head:** Always return the new node as the new head
4. **Simpler than tail:** Head insertion is easier and faster than tail insertion

---

## âš ï¸ **Common Pitfalls**

- **Not returning new_node:** Must return the new node as the new head
- **Wrong linking order:** Set `new_node.next = llist` BEFORE returning
- **Forgetting to update head:** The caller must update their head pointer with the return value
- **Confusing with tail insertion:** Head insertion is simpler - no traversal needed!
