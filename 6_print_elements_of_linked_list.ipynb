## ğŸ“ Linked List Node Structure

A linked list node contains:
- **data:** The value stored in the node
- **next:** Pointer to the next node (None for the last node)


````python
class SinglyLinkedListNode:
    def __init__(self, node_data):
        self.data = node_data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def insert_node(self, node_data):
        node = SinglyLinkedListNode(node_data)
        if not self.head:
            self.head = node
        else:
            self.tail.next = node
        self.tail = node
````

## ğŸ“¥ Input Format

- First line: `n` (number of elements)
- Next `n` lines: Each element value


````python
# Reading input and building linked list
llist_count = int(input())
llist = SinglyLinkedList()

for _ in range(llist_count):
    llist_item = int(input())
    llist.insert_node(llist_item)

print(f"Linked list created with {llist_count} nodes")
````

## ğŸ’¡ Solution: Linked List Traversal

**Time Complexity:** O(n) - Visit each node once  
**Space Complexity:** O(1) - Only use a pointer variable

The key is to use a current pointer that moves through the list until it reaches None (end of list).


````python
def printLinkedList(head):
    """
    Print all elements of a linked list.
    
    Args:
        head: Pointer to the head node of the linked list
    """
    current = head
    while current is not None:
        print(current.data)
        current = current.next  # Move to next node

# Test the function
print("\nPrinting linked list:")
printLinkedList(llist.head)
````

## ğŸ” **Step-by-Step Walkthrough**

For linked list `16 -> 13 -> NULL`:

1. **Initial:** `current = head` (points to node with data 16)
2. **Iteration 1:** Print 16, move `current = current.next` (now points to node with data 13)
3. **Iteration 2:** Print 13, move `current = current.next` (now points to None)
4. **Stop:** `current is None`, exit loop

---

## ğŸ” **Complexity Analysis**

- **Time Complexity:** O(n) - We visit each node exactly once
- **Space Complexity:** O(1) - We only use a constant amount of extra space (the `current` pointer)

---

## ğŸ¯ **Key Takeaways**

1. **Traversal pattern:** `current = head; while current is not None; current = current.next`
2. **Always check for None:** The loop condition `current is not None` prevents accessing None.next
3. **Preserve head:** Don't modify the head pointer directly - use a current pointer
4. **Foundation skill:** This traversal pattern is used in almost every linked list problem

---

## âš ï¸ **Common Pitfalls**

- **Modifying head:** Don't change `head` directly - use a `current` variable
- **Wrong loop condition:** Use `current is not None` not `current.next is not None` (misses last node)
- **Forgetting to move pointer:** Always update `current = current.next` or you'll have an infinite loop
- **Accessing None:** Check `current is not None` before accessing `current.data` or `current.next`
