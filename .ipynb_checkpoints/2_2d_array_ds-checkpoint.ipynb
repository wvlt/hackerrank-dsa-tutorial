## ðŸ“¥ Input Format

````python
# Reading input - 6x6 2D array
arr = []

for _ in range(6):
    arr.append(list(map(int, input().rstrip().split())))

print("2D Array:")
for row in arr:
    print(row)
````

## ðŸ’¡ Understanding the Hourglass Pattern

An hourglass in a 2D array consists of 7 elements arranged as:

```
arr[i][j]     arr[i][j+1]     arr[i][j+2]
              arr[i+1][j+1]
arr[i+2][j]   arr[i+2][j+1]   arr[i+2][j+2]
```

**Key Insight:** For a 6Ã—6 array, we can have hourglasses starting at positions (0,0) through (3,3), giving us 4Ã—4 = 16 possible hourglasses.


## ðŸ’¡ Solution Approach: Nested Loop with Pattern Recognition

**Time Complexity:** O(1) - Always processes 16 hourglasses regardless of input size  
**Space Complexity:** O(1) - Only uses a few variables

The approach:
1. Iterate through all possible starting positions (i, j) where an hourglass can fit
2. Calculate the sum of each hourglass pattern
3. Track the maximum sum encountered


````python
def hourglassSum(arr):
    """
    Find the maximum hourglass sum in a 6x6 2D array.
    
    An hourglass pattern:
    a b c
      d
    e f g
    
    Args:
        arr: 6x6 2D list of integers
        
    Returns:
        Maximum hourglass sum
    """
    max_sum = float('-inf')  # Initialize with negative infinity to handle negative sums
    
    # We can start hourglasses from (0,0) to (3,3)
    # This gives us 4x4 = 16 possible hourglasses
    for i in range(4):
        for j in range(4):
            # Calculate hourglass sum
            # Top row: arr[i][j], arr[i][j+1], arr[i][j+2]
            top = arr[i][j] + arr[i][j + 1] + arr[i][j + 2]
            
            # Middle element: arr[i+1][j+1]
            middle = arr[i + 1][j + 1]
            
            # Bottom row: arr[i+2][j], arr[i+2][j+1], arr[i+2][j+2]
            bottom = arr[i + 2][j] + arr[i + 2][j + 1] + arr[i + 2][j + 2]
            
            current_sum = top + middle + bottom
            
            # Update maximum if current sum is greater
            if current_sum > max_sum:
                max_sum = current_sum
    
    return max_sum

# Test the function
result = hourglassSum(arr)
print(f"\nMaximum hourglass sum: {result}")
````

## ðŸ’¡ Alternative Solution: Single Expression

This version calculates the sum in one line, which can be more concise but slightly less readable.


````python
def hourglassSum(arr):
    """
    Alternative implementation using single expression.
    """
    max_sum = float('-inf')
    
    for i in range(4):
        for j in range(4):
            # Calculate all 7 elements in one expression
            current_sum = (arr[i][j] + arr[i][j + 1] + arr[i][j + 2] + 
                          arr[i + 1][j + 1] + 
                          arr[i + 2][j] + arr[i + 2][j + 1] + arr[i + 2][j + 2])
            
            if current_sum > max_sum:
                max_sum = current_sum
    
    return max_sum

# Test alternative solution
result_alt = hourglassSum(arr)
print(f"Maximum hourglass sum (alternative): {result_alt}")
````

## ðŸ” Complexity Analysis

- **Time Complexity:** O(1) - We always iterate through exactly 16 hourglasses (4Ã—4), regardless of the array size
- **Space Complexity:** O(1) - We only use a constant amount of extra space (max_sum variable)

**Note:** While this is technically O(1) for a fixed-size 6Ã—6 array, if we generalize to nÃ—n arrays, the complexity would be O(nÂ²).

## ðŸŽ¯ Key Takeaways

1. **Pattern Recognition:** Identifying the hourglass pattern is crucial - it's a fixed 3Ã—3 pattern with a gap in the middle
2. **Boundary Conditions:** For a 6Ã—6 array, valid starting positions are (0,0) through (3,3)
3. **Initialization:** Use `float('-inf')` instead of 0 to handle arrays with all negative values
4. **Index Management:** Careful attention to indices prevents off-by-one errors

## âš ï¸ Common Pitfalls

- **Wrong range:** Using `range(6)` instead of `range(4)` will cause index out of bounds
- **Initializing max_sum to 0:** This fails when all hourglasses have negative sums
- **Missing elements:** Forgetting one of the 7 elements in the hourglass pattern
- **Index confusion:** Mixing up row and column indices (i vs j)
