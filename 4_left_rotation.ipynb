## ğŸ“¥ Input Format

- First line: `n` (number of integers) and `d` (number of left rotations)
- Second line: `n` space-separated integers


```python
# Reading input
first_multiple_input = input().rstrip().split()
n = int(first_multiple_input[0])  # Number of elements
d = int(first_multiple_input[1])  # Number of left rotations

arr = list(map(int, input().rstrip().split()))

print(f"Array length: {n}")
print(f"Rotations: {d}")
print(f"Original array: {arr}")
```

## ğŸ’¡ Solution Approach 1: Array Slicing (Pythonic)

**Time Complexity:** O(n) - Creates new array  
**Space Complexity:** O(n) - Creates new array

This is the most Pythonic and efficient approach for Python. It uses slicing to create a new rotated array.


```python
def rotateLeft(d, arr):
    """
    Rotate array left by d positions using array slicing.
    
    Args:
        d: Number of left rotations
        arr: List of integers
        
    Returns:
        Rotated list
    """
    # Take elements from index d to end, then from start to d
    return arr[d:] + arr[:d]

# Test the function
result = rotateLeft(d, arr.copy())
print(f"Rotated array (slicing): {result}")
```

## ğŸ’¡ Solution Approach 2: One-by-One Rotation (In-Place)

**Time Complexity:** O(n Ã— d) - Performs d rotations, each taking O(n)  
**Space Complexity:** O(1) - Modifies array in-place

This approach rotates one element at a time, modifying the array in-place. While less efficient, it demonstrates the concept clearly.


```python
def rotateLeft(d, arr):
    """
    Rotate array left by d positions using one-by-one rotation (in-place).
    
    Args:
        d: Number of left rotations
        arr: List of integers (will be modified)
        
    Returns:
        The same list, rotated in-place
    """
    for _ in range(d):
        first = arr.pop(0)  # Remove first element
        arr.append(first)   # Append to end
    
    return arr

# Test the function
test_arr = arr.copy()
result = rotateLeft(d, test_arr)
print(f"Rotated array (one-by-one): {result}")
print(f"Original array modified: {test_arr}")
```

## ğŸ” **Complexity Analysis**

### Approach 1: Array Slicing
- **Time:** O(n) - Creates new array by concatenating two slices
- **Space:** O(n) - Creates new array
- **Best for:** When you need a new rotated array and want clean, readable code

### Approach 2: One-by-One Rotation
- **Time:** O(n Ã— d) - Performs d rotations, each taking O(n) due to pop(0)
- **Space:** O(1) - Modifies array in-place
- **Best for:** When you need in-place modification (though slicing is still preferred in Python)

**Note:** In Python, `pop(0)` is O(n) because it requires shifting all elements. For large arrays and many rotations, slicing is significantly faster.

---

## ğŸ¯ **Key Takeaways**

1. **Array slicing** (`arr[d:] + arr[:d]`) is the most efficient Python approach
2. **One-by-one rotation** demonstrates the concept but is slower due to `pop(0)` overhead
3. For Python, slicing is almost always preferred unless you specifically need in-place modification
4. Understanding both approaches helps in interviews and different language contexts

## âš ï¸ **Common Pitfalls**

- **Using d rotations when d > n:** You can optimize by using `d % n` to avoid unnecessary rotations
- **Modifying original array:** If you need to preserve the original, make a copy first
- **Index errors:** Ensure d doesn't exceed array length (or use modulo)
- **Performance:** Using `pop(0)` in a loop is inefficient for large arrays
