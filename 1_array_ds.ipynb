## ğŸ“¥ Input Format

The first line contains an integer `n`, the number of elements in the array.
The second line contains `n` space-separated integers.


```python
# Reading input
arr_count = int(input().strip())
arr = list(map(int, input().rstrip().split()))

print(f"Array length: {arr_count}")
print(f"Original array: {arr}")
```

## ğŸ’¡ Solution Approach 1: Python Slicing (Simplest)

**Time Complexity:** O(n)  
**Space Complexity:** O(n) - creates a new array

This is the most Pythonic way to reverse an array. While elegant, it creates a new array rather than modifying in-place.


```python
def reverseArray(arr):
    """
    Reverse an array using Python slicing.
    
    Args:
        arr: List of integers
        
    Returns:
        Reversed list
    """
    return arr[::-1]

# Test the function
result = reverseArray(arr.copy())
print(f"Reversed array (slicing): {result}")
```

## ğŸ’¡ Solution Approach 2: Two-Pointer Technique (In-Place)

**Time Complexity:** O(n)  
**Space Complexity:** O(1) - modifies array in-place

This approach uses two pointers starting from both ends of the array, swapping elements until they meet in the middle. This is the preferred method when you need to modify the array in-place and is commonly asked in interviews.


```python
def reverseArray(arr):
    """
    Reverse an array using two-pointer technique (in-place).
    
    Args:
        arr: List of integers (will be modified)
        
    Returns:
        The same list, reversed in-place
    """
    left = 0
    right = len(arr) - 1
    
    # Swap elements until pointers meet
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1
    
    return arr

# Test the function
test_arr = arr.copy()
result = reverseArray(test_arr)
print(f"Reversed array (two-pointer): {result}")
print(f"Original array modified: {test_arr}")
```

## ğŸ” **Complexity Analysis**

### Approach 1: Python Slicing
- **Time:** O(n) - must copy all elements
- **Space:** O(n) - creates new array
- **Best for:** When you need a new reversed array and want clean, readable code

### Approach 2: Two-Pointer
- **Time:** O(n) - visits each element once
- **Space:** O(1) - only uses constant extra space
- **Best for:** When you need in-place modification or want to demonstrate algorithmic thinking

## ğŸ¯ **Key Takeaways**

1. **Python slicing** (`[::-1]`) is elegant but creates a new array
2. **Two-pointer technique** is space-efficient and demonstrates strong algorithmic skills
3. Both approaches have O(n) time complexity, but differ in space usage
4. In interviews, the two-pointer method often shows deeper understanding
